```python
df.dropna(inplace=True)
```

```python
tfidf = TfidfVectorizer(stop_words='english')

matrix = tfidf.fit_transform(df['description'])

matrix.shape
```

```python
print(matrix)
```

```python
lin_sim = linear_kernel(matrix,matrix)
print(lin_sim)
```

```python
# using cosine similarity
cos_sim = cosine_similarity(matrix)
print(cos_sim)
```

```python
sim_df = pd.DataFrame(cos_sim, index=df['title'], columns=df['title'])
print(sim_df)
```

```python
ask_movie = input('Hey there , Enter movie you like:')
print(ask_movie)
```

```python
movie_index = sim_df.index.get_loc(ask_movie)


top_10 = sim_df.iloc[movie_index].sort_values(ascending=False)[1:11]
```

```python
print(f'Top 10 similar movies to {ask_movie}:')
print(top_10)
```
